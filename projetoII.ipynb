# _Equação usada:_

Lei de Coulomb (Eletricidade e Magnetismo) dada pela seguinte equação:
$F = \frac{k * q_1 * q_2}{r^2}$

Onde $F$ é a força eletrostática, $k$ é a constante eletrostática, $q1$ e $q2$  são as cargas das partículas e $r$ é a distância entre elas.

**Probela a ser resolvido:** Determine a distância entre duas cargas elétricas de módulo $5⋅10^{−3}$ C
que se repelem no vácuo com uma força elétrica de $20 N$.
Dado: $k=9⋅10^9 \frac{Nm2}{C2}$

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from scipy import optimize
from sympy.utilities.lambdify import lambdify
import scipy.optimize as optimize
```

```python
F = 20  # Força elétrica
q1 = q2 = 5e-3
k = 9e9

f = lambda r: np.sqrt((k * q1 * q2) / r**2)

t1 = np.linspace(0, 200, 30)
t2 = np.linspace(20, 100, 10)

fig, axs = plt.subplots(1, 2, figsize=(18, 9))  # Cria uma figura com 1 linha e 2 colunas de subplots

# Primeiro subplot
axs[0].plot(t1, f(t1), 'b--')
axs[0].set_title('Gráfico 1')
axs[0].set_xlabel("Distncia r")
axs[0].set_ylabel("Força N")
axs[0].grid()

# Segundo subplot
axs[1].plot(t2, f(t2), "g--")
axs[1].set_title('Gráfico 2')
#axs[1].set_yticks([0, 0.5, 1, 1.5, 2, 2.5])
axs[1].set_xlabel("Distncia r")
axs[1].set_ylabel("Força N")
axs[1].grid()

plt.tight_layout()  # Ajusta o espaçamento entre subplots
plt.show()
```

Podemos usar a derivada e as informações fornecidas para determinar a distância entre duas cargas elétricas que se repelem no vácuo com uma força elétrica de $20 N$. A fórmula para calcular a distância $r$ a parti da derivada da força $f'(r)$ é a seguinte:

$f'(r) = -2 * \frac{k* |q_1 * q_2|}{r^3}$

Dado que $f'(r) = \frac{df(r)}{dr}$, podemos igular a derivada f'(r) a fórmula acima para resolver $r$:

$-2 * \frac{k* |q_1 * q_2|}{r^3} = 20$

```python
# Símbolos
r, k, q1, q2 = sp.symbols('r k q1 q2', real=True)
# Fórmula da derivada
derivada_formula = -2 * (k * abs(q1 * q2)) / r**3
# Equação a ser resolvida
equacao = sp.Eq(derivada_formula, 20)
# Resolve a equação para r
distancia_r = sp.solve(equacao, r)

# Lambdify para obter o valor numérico
distancia_r_numeric = lambdify((k, q1, q2), distancia_r, "numpy")
# Valores das constantes
k_value = 9e9  # Nm^2/C^2
q1_value = q2_value = 5e-3  # C
# Calcula as distâncias numéricas
distancias_numeric = distancia_r_numeric(k_value, q1_value, q2_value)

# Imprime uma tabela
print("Possíveis valores da distância:")
print("------------------------------")
print("Distância (metros)")
print("------------------")
for distancia in distancias_numeric:
    print(f"{distancia:.4f}")
```
