Como no DataFrame é constatado que não há a presença de alguns vinhos identificados pelo $Id$, pode-se calcular a estimativa da qualidade do vinho por meio da interpolação. Assim:
  
A interpolação linear é uma técnica matemática que envolve a estimativa de valores intermediários com base em pontos de dados conhecidos. No código de interpolação fornecido, a interpolação linear é realizada com as seguintes etapas, expressas em termos de fórmulas matemáticas:

1. *Dados de entrada:*
 
    $Id$: Vetor de identificadores (IDs) conhecidos.       
    $quality$: Vetor de valores de qualidade conhecidos.

2. *Criação da Função de Interpolação Linear:*

    Uma função de interpolação linear é criada usando a função _interp1d_ da biblioteca SciPy. Ela recebe os dados de entrada, $Id$ e $quality$, e especifica o tipo de interpolação como linear.    
    $f(x) = inter1d(Id, quality, kind='linear')$

3. *Previsão da Qualidade para um Novo ID:*

    Para prever a qualidade para um novo ID, é fornecido um novo valor como $X_{new}$ à função $f(x)$ criada: $quality_{pred} = f(x_{new})$    
    A função $f(x)$ estima a qualidade com base nos pontos de dadso conhecidos usando interpolação linear.

```python
from scipy.interpolate import interp1d

# Ordene o DataFrame pelo 'Id' para garantir que os dados estejam em ordem
# Extrai as colunas 'quality' e 'Id' como arrays NumPy
df = wine.sort_values(by='Id')
quality = df['quality'].values
Id = df['Id'].values

interp_func = interp1d(Id, quality, kind='linear', fill_value="extrapolate")
all_ids = df['Id'].values # Identifique os valores de 'Id' ausentes
missing_ids = [id for id in range(all_ids[0], all_ids[-1] + 1) if id not in all_ids]
predicted_qualities = pd.Series([interp_func(new_id) for new_id in missing_ids], index=missing_ids)
predicted_qualities.head(15)
```

#### O código da previsão da qualidade do vinho usando regressão linear pode ser expresso em termos de fórmulas matemáticas da seguinte maneira:

Suponhamos que temos um conjunto de dados com *n* amostras e *m* variáveis independentes

* *X*: Matriz de características, com dimensões $n x (m + 1)$ nde a primeira coluna contém 1s (para o termo de viés) e as colunas restantes representam as características.      
* *y*: Vetor de rótulos de qualidade, com dimensões $n + 1$.

O objetivo é encontrar um vetor de coeficientes *w* de dimensões $(m + 1) x 1$ que minimize o erro da regressão. Isso é feito resolvendo o sistema de equações normais.

1. *Modelo de Regressão Linear:*
    A fórmula do modelo de regressão linear múltipla é dada por:             
    $Y = X_w$, onde:        
        - *y* é o vetor de rótulos              
        - *X* é a matriz de características      
        - *w*  é o vetor de coeficientes que  estamos tentando encontrar.

2. *Sistema de Equações Normais:*
    O sistema de equações normais é usado para encontrar o vetor de coeficientes *w* que minimiza o erro da regressão. Ele é dado por: 
    $X^TX_w = X^T_{y}$, onde:        
        - $X^T$ é a matrix transposta de $X$             
        - $X^TX$ a matriz resultante da multiplicação matricial de $X^T$ com $X$     
        - $X^T_{y}$ é a matriz resultante da multiplicação matricial de $X^T$ com $y$              

3. *Solução do Sistema de Equações Normais:*
    A solução para *w* é encontrada usando a função _solve_ da biblioteca SciPy. O sistema de equações é resolvido da seguinte forma:
    $w = solve(X^TX, X^T_{y})$

4. *Previsão de Qualidade do Vinho:*
    Uma vez que *w* é encontrado, podemos usar esse vetor de coeficientes para fazer previsões de qualidade do vinho para novos dados de características $X_{pred}$. As previsões são dadas por:           
    $y_{pred} = X_{pred^w}$         
    onde, $X_{pred}$ é a matriz de características dos novos dados para os quais queremos fazer previsões, e $y_{pred}$ é o vetor de previsões de qualidade do vinho.

Portanto, o código realiza esses cálculos matemáticos para encontrar os coeficientes que melhor se ajustam aos dados de treinamento e, em seguida, usa esses coeficientes para fazer previsões da qualidade do vinho para os dados de teste ou quaisquer novos dados fornecidos. O objetivo é minimizar o erro da regressão e criar um modelo que capture as relações entre as características, as quais são indicativas de qualidade e a qualidade do vinho em si.

```python
import numpy as np
import zlib
from scipy.linalg import solve

'''
Criação de uma cópia do DataFrame original wine chamada wine_copy,
isso é feito para que as alterações não afetem o DataFrame original. 
Remoção da coluna 'quality' do DataFrame wine_copy e a conversão em um array NumPy chamado y,
a fim de ser o rótulo das previsões.
Remoção da coluna 'quality' do DataFrame wine_copy e conversão do restante dos dados 
em um array NumPy chamado X, o qual representa as variáveis independentes.
'''
wine_copy = wine.copy()
y = wine_copy['quality'].values
X = wine_copy.drop(columns=['quality']).values

# Geração de uma semente usando zlib.crc32 para garantir uma separação estável
seed = zlib.crc32(b'seed') & 0xffffffff

'''
Calculo do tamanho do conjunto de treinamento com 80% dos dados.
Use-se a semente para dividir os dados em treinamento e teste de maneira estável.
Divide o conjunto de dados em X_train (conjunto de treinamento) e X_test (conjunto de teste) 
com base no tamanho calculado.
Divão do rótulo y, em y_train e y_test para treinamento e teste.
'''
np.random.seed(seed)
indices = np.arange(len(X))
np.random.shuffle(indices)
train_size = int(0.8 * len(X))
train_indices = indices[:train_size]
test_indices = indices[train_size:]

X_train, X_test = X[train_indices], X[test_indices]
y_train, y_test = y[train_indices], y[test_indices]

# Adiciona uma coluna de 1s no início dos dados de treinamento. 
# Essa coluna de 1s é usada para o termo de viés na regressão linear.
X_train = np.hstack((np.ones((X_train.shape[0], 1)), X_train))
X_test = np.hstack((np.ones((X_test.shape[0], 1)), X_test))

# Realizar a regressão linear usando solve
# Calcula as previsões da qualidade do vinho com base nos dados 
# de teste usando o vetor de coeficientes w
w = solve(X_train.T @ X_train, X_train.T @ y_train)
y_pred = X_test @ w

# Calcule as previsões da qualidade do vinho para o DataFrame wine_copy usando os coeficientes w
X_pred = wine_copy.drop(columns=['quality']).values
X_pred = np.hstack((np.ones((X_pred.shape[0], 1)), X_pred))
qualidade_predita = X_pred @ w

wine_copy['Qualidade_Prevista'] = qualidade_predita
wine_copy[730:750]
```

```python
# Calcular o Erro Médio Absoluto (MAE)
mae = np.mean(np.abs(y_pred - y_test))

# Calcular o Erro Quadrático Médio (MSE)
mse = np.mean((y_pred - y_test) ** 2)

# Calcular o erro relativo
erro_relativo = mae / np.mean(y_test)

# Calcular a acurácia usando MAE como limite
limite_acuracia = 0.95  # intervalo de confiança para o erro de generalização
acuracia = np.mean(np.abs(y_pred - y_test) <= limite_acuracia)

print(f"Erro Médio Absoluto (MAE): {mae}")
print(f"Erro Quadrático Médio (MSE): {mse}")
print(f"Erro Relativo: {erro_relativo}")
print(f"Acurácia: {acuracia * 100}%")
```
